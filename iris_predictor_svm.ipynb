```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()

dir(iris)
```

```python
iris.feature_names
```

```python
import pandas as pd

df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df["target"]=iris.target
df.head()
```

```python
iris.target_names
```

```python
df["flower name"]=df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
import matplotlib.pyplot as plt
df0=df[iris.target==0]
df1=df[iris.target==1]
df2=df[iris.target==2]

plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color="red",marker="*")
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color="blue",marker="+")
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color="green",marker=".")
```

```python
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color="red",marker="*")
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color="blue",marker="+")
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],color="green",marker=".")
```

```python
x=df.drop(["target","flower name"], axis="columns")
x.head()
```

```python
y=df.target
y.head()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2)

```

```python
from sklearn.svm import SVC

model=SVC()

model.fit(x_train,y_train)

```

```python
model.score(x_test,y_test)
```
